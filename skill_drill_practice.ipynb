```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the books to Scrape site
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('a').text
title
```

```python
# Scrape the book name
tag_box = html_soup.find_all('article', class_='product_pod')
for t in tag_box:
    # tag_box
    tags = t.find_all('a')

    for tag in tags:
        word = tag.text
        print(word)
```

```python
for x in range(1, 6):
   html = browser.html
   quote_soup = soup(html, 'html.parser')
   quotes = quote_soup.find_all('article', class_='product_pod')
   #quotes = quote_soup.find_all('div', class_='quote')  #question of what would happen
   for quote in quotes:
      print('page:', x, '----------')
      print(quote.text)
   browser.links.find_by_partial_text('Next').click()
```
